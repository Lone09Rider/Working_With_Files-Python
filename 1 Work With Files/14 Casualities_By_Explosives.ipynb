```python
import csv
```

```python
sum = 0
killed = []
wounded = []
with open("year2017.csv") as file:
    data = csv.DictReader(file, skipinitialspace=True)
    
    for row in data:
        if row["Weapon_type"] == "Explosives":
            killed.append(row["Killed"])
            wounded.append(row["Wounded"])
```

```python
new_killed = []
for i in killed:
    if i != "":
        new_killed.append(float(i))
new_killed
```

```python
new_wounded = []
for i in wounded:
    if i != "":
        new_wounded.append(float(i))
new_wounded
```

```python
wounded_sum = killed_sum = 0
for i in new_killed:
     killed_sum += i
```

```python
killed_sum
```

```python
for i in new_wounded:
     wounded_sum += i
```

```python
wounded_sum
```

```python
Casualities_By_Explosives = killed_sum + wounded_sum
```
