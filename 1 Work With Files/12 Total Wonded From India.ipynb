```python
import csv
```

```python
with open("year2017.csv") as file:
    data = list(csv.reader(file))
```

```python
data[0]
```

```python
wounded = []
for row in data[:]:
    if row[3] == "India":
        wounded.append(row[10])
```

```python
wounded
```

```python
indian_wounded = []
for i in wounded:
    if i != "":
        indian_wounded.append(float(i))
```

```python
indian_wounded
```

```python
sum = 0
for i in indian_wounded:
    sum += i
```
