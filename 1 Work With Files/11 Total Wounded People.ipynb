```python
import csv
```

```python
with open("year2017.csv") as file:
    data = list(csv.reader(file))
```

```python
data[0]
```

```python
wounded = []
```

```python
for row in data[1:]:
    wounded.append(row[10])
```

```python
wounded
```

```python
new_wounded = []
for i in wounded:
    if i != "":
        new_wounded.append(float(i))
```

```python
new_wounded
```

```python
sum = 0
for i in new_wounded:
    sum += i
```
