```python
import csv
```

```python
with open("year2017.csv") as file_obj:
    data = list(csv.DictReader(file_obj))

Country_Wise_Killed = {}
for row in data:
    key = row["Country"]
    value = row["Killed"]
    
    if value != "":
        value = int(float(value))
    
    else:
        value = 0
    
    if key in Country_Wise_Killed:
        Country_Wise_Killed[key] += value
    else:
        Country_Wise_Killed[key] = value
```
